# Visualization 1 : Density of features.

```python
data.plot(kind = "density", figsize=(10,8))
```

## Visualization 2: comparison between various classifications based on SL and SW

```python
data['Classification'].value_counts()
```

```python
fig, ax = plt.subplots()

colors = {'Iris-versicolor':'red', 'Iris-virginica':'green', 'Iris-setosa':'blue'}
plt.title(r'comparison between various classification based on SL and SW')
grouped = data.groupby('Classification')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='SL', y='SW', label=key, color=colors[key] ,s=25)
plt.show()
```

## Visualization 3:

```python
sns.pairplot(data,hue='Classification',height=4)
```

### Distribution of 'classification' with all other features.

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.boxplot( y="PW", x= "Classification", data=data, orient='v' , ax=axes[0, 0])
sns.boxplot( y="PL", x= "Classification", data=data, orient='v' , ax=axes[0, 1])
sns.boxplot( y="SL", x= "Classification", data=data, orient='v' , ax=axes[1, 0])
sns.boxplot( y="SW", x= "Classification", data=data, orient='v' , ax=axes[1, 1])
plt.show()
```
